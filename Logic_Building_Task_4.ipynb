## Problem 1: Mobile Recharge Validation System

This program validates a mobile recharge request.
Recharge amount must be at least ₹50 and must match valid plans.
The user can retry if the input is invalid.


```python
# Function to validate recharge
def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: "))

        # Check minimum recharge rule
        if amount < 50:
            print("Recharge must be at least ₹50. Try again.")
        # Check valid plan
        elif amount not in valid_plans:
            print("Invalid plan. Available plans:", valid_plans)
        else:
            print("Recharge successful for ₹", amount)
            break

# Call the function
validate_recharge()
```

## Problem 2: Inventory Reorder Alert System

This program checks product stock levels and prints a reorder alert
if stock is less than 15.


```python
# Dictionary of products and stock quantities
inventory = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 12,
    "Monitor": 18
}

# Function to check stock
def check_inventory(stock_data):
    for product, quantity in stock_data.items():
        if quantity < 15:
            print(product, "→ Reorder Alert")
        else:
            print(product, "→ Stock OK")

# Call the function
check_inventory(inventory)
```

## Problem 3: Student Result Processing System

This program calculates the average marks of students
and determines whether the result is Pass or Fail.


```python
# Function to process student results
def check_result(marks):
    total = 0

    # Calculate total marks
    for mark in marks:
        total += mark

    # Calculate average
    average = total / len(marks)

    print("Average Marks:", average)

    # Check pass or fail
    if average >= 50:
        return "Pass"
    else:
        return "Fail"

# Sample marks
student_marks = [60, 55, 70, 45]

# Display result
result = check_result(student_marks)
print("Result:", result)
```

## Problem 4: Cab Fare Estimator

This program calculates cab fare based on distance
and adds extra charges during peak hours.
User can retry the calculation.


```python
# Function to calculate cab fare
def calculate_fare(distance, peak):
    base_fare = 50
    per_km_rate = 12

    fare = base_fare + (distance * per_km_rate)

    # Add peak hour charges
    if peak.lower() == "yes":
        fare += fare * 0.25

    return fare

# Retry system
while True:
    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ")

    total_fare = calculate_fare(distance, peak)
    print("Total Fare: ₹", total_fare)

    retry = input("Do you want to calculate again? (yes/no): ")
    if retry.lower() != "yes":
        break
```

## Problem 5: Employee Attendance Eligibility Checker

This program checks whether an employee is eligible
based on attendance percentage.


```python
# Function to check attendance eligibility
def check_attendance(attendance_list):
    present_days = 0

    # Count present days
    for status in attendance_list:
        if status == "P":
            present_days += 1

    percentage = (present_days / len(attendance_list)) * 100
    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

# Example attendance
attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]

# Display result
status = check_attendance(attendance)
print("Status:", status)
```

## Problem 6: Password Strength Checker

This program checks whether a password is strong
based on length, digits, and special characters.


```python
# Function to check password strength
def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"

    # Check conditions
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    if len(password) >= 8 and has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"

# Input password
password = input("Enter password: ")

# Display result
print(check_password(password))
```
