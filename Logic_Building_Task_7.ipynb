## Problem 1: Smart Light Controller

Create a class to control a smart light that can be turned ON or OFF
and display its current status.

```python
# Smart Light Controller Class
class SmartLight:
    def __init__(self, light_name):
        # Initialize light name
        self.light_name = light_name
        self.status = "OFF"

    def turn_on(self):
        self.status = "ON"
        print(self.light_name, "is", self.status)

    def turn_off(self):
        self.status = "OFF"
        print(self.light_name, "is", self.status)

# Sample Input
light = SmartLight("Bedroom Light")
light.turn_on()
```

## Problem 2: Employee ID Card System

Create a class that stores employee name, ID, and department
and displays ID card details.

```python
# Employee ID Card Class
class Employee:
    def __init__(self, name, emp_id, department):
        # Store employee details
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)

# Sample Input
emp = Employee("Rahul", "EMP102", "IT")
emp.display_id_card()
```

## Problem 3: Mobile Contact Record

Create a class to store contact name and phone number
and display contact information.

```python
# Contact Record Class
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)

# Sample Input
contact = Contact("Anita", "9876543210")
contact.display_contact()
```

## Problem 4: Product Price Tag Generator

Create a class that stores product name and price
and prints a formatted price tag.

```python
# Product Price Tag Class
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def display_price_tag(self):
        print("Product:", self.name)
        print("Price: ₹", self.price)

# Sample Input
product = Product("Headphones", 2499)
product.display_price_tag()
```

## Problem 5: Movie Rating Display System

Create a class that stores movie name and rating
and displays movie details.

```python
# Movie Rating Class
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display_movie(self):
        print("Movie:", self.name)
        print("Rating:", self.rating, "/ 5")

# Sample Input
movie = Movie("Inception", 4.8)
movie.display_movie()
```

## Problem 6: Delivery Address Manager

Create a class that stores customer name and delivery address
and prints delivery details.

```python
# Delivery Address Class
class Delivery:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def display_delivery(self):
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)

# Sample Input
delivery = Delivery("Suman", "Hyderabad")
delivery.display_delivery()
```
