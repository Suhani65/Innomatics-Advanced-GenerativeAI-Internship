```python
#1 User Login Check

username = "admin"
password = "1234"
if username == "admin" and password == "1234":
  print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
#2 Pass / Fail Analyzer

marks = [45, 78, 90, 33, 60]
pass_count = 0
fail_count = 0
for mark in marks:
  if mark>=50:
    pass_count += 1
  else:
    fail_count +=1
print("Passed:", pass_count)
print("Failed:", fail_count)
```

```python
#3 Simple Data Cleaner

names = [" Alice ", "bob", " CHARLIE "]
cleaned_names = []
for name in names:
  cleaned_names.append(name.strip().lower())
print("Cleaned Names:", cleaned_names)
```

```python
#4 Message Length Analyzer

messages = ["Hi", "Welcome to the platform", "OK"]
for msg in messages:
  print("Message:", msg)
  print("Length:", len(msg))
  if len(msg) > 10:
    print("Flag: Message longer than 10 characters")
  print("----")
```

```python
#5 Error Message Detector

logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count = logs.count("ERROR")
print("Total ERROR messages:", error_count)
```
