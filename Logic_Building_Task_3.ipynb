## Problem 1: Employee Performance Bonus Eligibility

This program identifies employees with the highest performance score.
If multiple employees have the same highest score, all are displayed.


```python
# Dictionary containing employee names and their performance scores
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Step 1: Find the highest performance score
highest_score = max(employees.values())

# Step 2: Identify employees who have the highest score
top_performers = []

for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)

# Step 3: Display the result
print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {highest_score})")
```

## Problem 2: Search Query Keyword Analysis

This program converts a search query to lowercase, removes punctuation,
counts word frequency, and displays words appearing more than once.


```python
import string

# Input search query
query = "Buy mobile phone buy phone online"

# Step 1: Convert to lowercase
query = query.lower()

# Step 2: Remove punctuation
query = query.translate(str.maketrans('', '', string.punctuation))

# Step 3: Split sentence into words
words = query.split()

# Step 4: Count frequency of each word
word_count = {}

for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

# Step 5: Display words searched more than once
repeated_words = {}

for word, count in word_count.items():
    if count > 1:
        repeated_words[word] = count

print(repeated_words)
```

## Problem 3: Sensor Data Validation

This program identifies valid sensor readings (even numbers)
and stores them as (hour_index, reading_value) pairs.


```python
# List of sensor readings
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# List to store valid readings
valid_readings = []

# Loop through readings with index
for index, value in enumerate(sensor_readings):
    if value % 2 == 0:  # Check if even (valid reading)
        valid_readings.append((index, value))

# Display result
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

## Problem 4: Email Domain Usage Analysis

This program counts how many users belong to each email domain
and calculates the percentage usage of each domain.


```python
# List of email IDs
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Dictionary to store domain counts
domain_count = {}

# Count domains
for email in emails:
    domain = email.split("@")[1]

    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

# Calculate and display percentage usage
total_emails = len(emails)

for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {percentage:.0f}%")
```

## Problem 5: Sales Spike Detection

This program calculates average daily sales
and detects days where sales exceed 30% above average.


```python
# List of daily sales
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Step 1: Calculate average sales
average_sales = sum(sales) / len(sales)

# Step 2: Detect sales spikes (more than 30% above average)
threshold = average_sales * 1.30

for day, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {day}: {value}")
```

## Problem 6: Duplicate User ID Detection

This program identifies duplicate user IDs
and displays how many times each duplicate appears.


```python
# List of user IDs
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Dictionary to store frequency
user_count = {}

# Count occurrences
for user in user_ids:
    if user in user_count:
        user_count[user] += 1
    else:
        user_count[user] = 1

# Display duplicates
for user, count in user_count.items():
    if count > 1:
        print(f"{user} → {count} times")
```
