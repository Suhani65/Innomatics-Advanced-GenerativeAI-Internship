## Problem 1: Smart Parking Lot Management System

This system manages parking entry and exit logs,
calculates currently parked vehicles,
and alerts if parking exceeds capacity.

```python
# Function to manage parking lot
def parking_system(capacity, logs):
    parked = 0

    # Process vehicle logs
    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT":
            parked -= 1

    print("Currently Parked Vehicles:", parked)

    # Check parking status
    if parked > capacity:
        print("Parking Status: Full")
    else:
        print("Parking Status: Available")

# Sample input
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

parking_system(capacity, vehicle_logs)
```

## Problem 2: Online Food Delivery Time Estimator

This system estimates delivery time based on:
- Distance
- Traffic Level
- Weather Conditions

```python
# Function to estimate delivery time
def delivery_time(distance, traffic, weather):
    time = distance * 5  # base time per km

    # Traffic delay
    if traffic == "High":
        time += 10
    elif traffic == "Medium":
        time += 5

    # Weather delay
    if weather == "Rainy":
        time += 5

    print("Estimated Delivery Time:", time, "minutes")

delivery_time(8, "High", "Rainy")
```

## Problem 3: Movie Theatre Seat Occupancy Analyzer

This system calculates occupancy percentage
and determines the show status.

```python
# Function to analyze occupancy
def theatre_occupancy(total_seats, booked_seats):
    occupancy = (len(booked_seats) / total_seats) * 100

    print("Occupancy:", int(occupancy), "%")

    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 70:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Available")

# Example booked seats
booked = [1] * 150

theatre_occupancy(200, booked)
```

## Problem 4: Cloud Server Load Classification System

This system calculates average CPU usage
and classifies server load.

```python
# Function to classify server load
def server_load(cpu_readings):
    total = 0

    for cpu in cpu_readings:
        total += cpu

    average = total / len(cpu_readings)

    print("Average CPU Load:", int(average), "%")

    if average < 50:
        print("Server Status: Normal")
    elif average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")

server_load([45, 60, 70, 85, 90])
```

## Problem 5: Smart Classroom Resource Usage Monitor

This system tracks usage of classroom resources
and identifies overused resources.

```python
# Function to monitor resource usage
def resource_monitor(resources):
    overused = []

    for resource, hours in resources.items():
        if hours > 8:
            overused.append(resource)

    print("Overused Resources:", ", ".join(overused))

    if overused:
        print("Energy Alert: Yes")
    else:
        print("Energy Alert: No")

# Sample input
usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

resource_monitor(usage)
```

## Problem 6: Online Event Registration Capacity Controller

This system manages event registrations,
prevents overbooking,
and triggers waitlist mode.

```python
# Function to manage registrations
def event_registration(capacity, registrations):
    if registrations > capacity:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlisted = 0
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)

event_registration(100, 105)
```
