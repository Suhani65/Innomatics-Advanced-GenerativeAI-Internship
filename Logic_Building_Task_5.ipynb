## Problem 1: Post Engagement Analyzer
Analyze total likes on a post and determine if it is viral.

```python
# Function to analyze post engagement
def analyze_engagement(likes):
    total_likes = 0

    # Calculate total likes
    for like in likes:
        total_likes += like

    print("Total Likes:", total_likes)

    # Check if viral
    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")

# Example data
likes_data = [200, 150, 300, 400]

analyze_engagement(likes_data)
```

## Problem 3: Agriculture – Rainfall Adequacy Checker

This program calculates the average rainfall and checks
whether it is adequate for agricultural needs.

```python
# Function to check rainfall adequacy
def rainfall_checker(rainfall_data, required_level):
    total = 0

    # Calculate total rainfall
    for rain in rainfall_data:
        total += rain

    average = total / len(rainfall_data)

    print("Average Rainfall:", average)

    if average >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")

rainfall_checker([70, 75, 80, 65], 70)
```

## Problem 4: Social Media – Duplicate Account Detection

This program detects duplicate usernames in a system.

```python
# Function to detect duplicate accounts
def detect_duplicates(usernames):
    unique_users = set(usernames)

    if len(unique_users) != len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

detect_duplicates(["user1", "user2", "user3", "user1"])
```

## Problem 6: Agriculture – Premium Crop Price Filter

This program filters crop prices that are above
the premium threshold.

## Problem 9: E-Learning – Student Attendance Eligibility System

This program calculates attendance percentage
and checks exam eligibility.

```python
# Function to check attendance
def attendance_checker(attendance):
    present = 0

    for day in attendance:
        if day == "P":
            present += 1

    percentage = (present / len(attendance)) * 100

    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")

attendance_checker(["P", "P", "A", "P", "P"])
```

## Problem 10: Smart Electricity Bill Analyzer

This program calculates electricity bill based on units consumed
and classifies usage level.

```python
# Function to calculate electricity bill
def electricity_bill(units):
    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    print("Total Bill:", bill)
    print("Usage Level:", status)

electricity_bill(250)
```
